```python
num = int(input("Enter your number:"))
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print("Your number is a prime number.")
else:
    print("Your number is not a prime number.")
```

```python
def is_prime(number):
    if number <= 1:
        return False
    if number <= 3:
        return True
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True

print("Prime numbers in the interval 1-10000:")
for num in range(1, 10001):
    if is_prime(num):
        print(num, end=' ')
```
